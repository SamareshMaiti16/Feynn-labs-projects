```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('cardiovascular-disease-death-rates-by-age.csv')
data.head()
```

```python
df=data.drop(['Entity', 'Code'],axis=1)
df.head()
```

```python
dfm=df.melt("Year",var_name="Age", value_name="Vals")
dfm.head()
```

```python
sns.lineplot(x="Year", y="Vals", hue='Age', data=dfm,legend='brief')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.show()
```
